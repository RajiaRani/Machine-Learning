```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Original XOR dataset in 2D
X = np.array([ [1, 1], [-1, -1], [1, -1], [-1, 1]])
y = np.array([1, 1, -1, -1])

# Plotting the 2D Points
plt.figure(figsize=(8,4))
for i, label in enumerate(y):
    plt.plot(X[i, 0], X[i,1], 'ro' if label==1 else 'bo')

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Original XOR points in 2D')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Original XOR dataset in 2D
X = np.array([ [1, 1], [-1, -1], [1, -1], [-1, 1]])
y = np.array([1, 1, -1, -1])

# Applying the feature transformation to 3D: x3 = x1 * x2 
X_tranformed = np.hstack((X, (X[:,0] * X[:,1]).reshape(-1, 1)))

fig = go.Figure()

colors = ['red' if label==1 else 'blue' for label in y]
fig.add_trace(go.Scatter3d(x=X_tranformed[:, 0], y=X_tranformed[:, 1], z=X_tranformed[:, 2] , mode='markers', marker=dict(color=colors, size=5)))


xx, yy = np.meshgrid(np.linspace(-2, 2, 50), np.linspace(-2, 2, 50))
zz = 0 * xx
fig.add_trace(go.Surface(x=xx, y=yy, z=zz, opacity=0.5, colorscale='Greens'))

fig.update_layout(title='Transformed XOR points in 3D with Separator', 
                  scene=dict(xaxis_title='x1', yaxis_title='x2', zaxis_title='x3=x1*x2'), margin=dict(l=0, r=0, b=0, t=30))

fig.show()
```
