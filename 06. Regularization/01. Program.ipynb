```python
from sklearn.datasets import make_classification
import pandas as pd

X, y = make_classification(
    n_samples=100,
    n_features=1,
    n_informative=1,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=42
)

df = pd.DataFrame({
    'X': X[:,0],   # feature column (like your Column B)
    'y': y         # label column (like your Column C)
})
df
```
