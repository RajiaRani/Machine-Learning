# What is Cross-Entropy Loss?

<p> When your model predicts probabilities (like 0.9 for cat, 0.1 for dog),
you need a way to measure how close those probabilities are to the true labels.</p>

<p>Thatâ€™s where Cross-Entropy Loss (also called Log Loss) comes in.</p>

<p>It measures how uncertain or wrong your predicted probability is.</p>

### ðŸ§® Binary Cross-Entropy Loss Formula

For binary classification, if:

- \( y = 1 \) â†’ actual label (true class)  
- \( \hat{y} \) â†’ predicted probability (from the sigmoid function)

Then the **Cross-Entropy Loss** is defined as:

\[
L = - \big[ y \cdot \log(\hat{y}) + (1 - y) \cdot \log(1 - \hat{y}) \big]
\]

### ðŸ’¡ Intuitive Explanation

| **True Label (y)** | **Predicted Probability (Å·)** | **Loss**                |
|---------------------|-------------------------------|--------------------------|
| 1                   | 0.9                           | small loss (good) âœ…     |
| 1                   | 0.1                           | large loss (bad) âŒ      |
| 0                   | 0.1                           | small loss (good) âœ…     |
| 0                   | 0.9                           | large loss (bad) âŒ      |

> **The better the prediction, the smaller the loss.**


```python
import numpy as np
```

```python
# define true labels and predicted probabilities
y_true = np.array([1, 0, 1, 0])       # actual (ground truth)
y_pred = np.array([0.9, 0.2, 0.8, 0.1])  # predicted probabilities from sigmoid
```

*What is happening?* <br/>
<p>The function np.clip() limits the values inside an array
to stay within a specified minimum and maximum range.</p>

<p>np.clip(array, min_value, max_value)</p>

```python
# compute cross-entropy loss
def binary_cross_entropy(y_true, y_pred):
    # small epsilon to avoid log(0)
    eps = 1e-15
    y_pred = np.clip(y_pred, eps, 1 - eps)
    loss = - np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
    return loss

loss = binary_cross_entropy(y_true, y_pred)
print("Cross-Entropy Loss:", loss)
```
