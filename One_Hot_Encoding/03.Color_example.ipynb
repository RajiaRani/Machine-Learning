*Computers can’t directly understand words like Red or Blue.* <br/>
*If we simply replace them with numbers (say Red = 1, Blue = 2, Green = 3),
the model will think 3 > 2 > 1, which is wrong, because colors don’t have order!* <br/>

*✅ One-Hot Encoding fixes this by creating new columns - one for each category -
and marking 1 where that category appears, 0 otherwise.*

```python
import pandas as pd
```

```python
# sample data 
data = pd.DataFrame({
    'Color' :['Red', 'Green', 'Blue', 'Red']
})

print('Original Data:')
print(data)
```

# Apply One-Hot Encoding

*Use pd.get_dummies() - a simple built-in function in pandas.*

```python
# One-Hot encode the 'Color' column
one_hot = pd.get_dummies(data['Color'], prefix='Color')

print('After OneHot Encoding: ')
print(one_hot)
```
