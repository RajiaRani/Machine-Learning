```python
import pandas as pd
```

```python
data = pd.DataFrame({
    'Country': ['India', 'USA', 'UK', 'India', 'France']
})
print('Original data : ',  data)
```

```python
OneHot_Encoding = pd.get_dummies(data, columns=['Country'])
print(OneHot_Encoding)
```

# ðŸ’¡ How to Get 0 and 1 Instead of True/False

*You can simply set dtype=int inside pd.get_dummies():*

```python
OneHot_Encoding = pd.get_dummies(data, columns=['Country'], dtype=int)
print(OneHot_Encoding)
```
