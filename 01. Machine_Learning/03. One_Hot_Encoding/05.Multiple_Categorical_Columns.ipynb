```python
import pandas as pd
```

```python
data = pd.DataFrame({
    'Gender' : ['Male','Female', 'Female', 'Male','Male'],
      'City': ['Delhi', 'Mumbai', 'Delhi', 'Bangalore', 'Mumbai']
})

print('Original Data: ', data)
```

```python
one_hot_encoded = pd.get_dummies(data, columns=['Gender', 'City'], dtype=int)
print(one_hot_encoded)
```
